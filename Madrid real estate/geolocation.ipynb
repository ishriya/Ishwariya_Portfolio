```python
import pandas as pd
import concurrent.futures
import time
import googlemaps
import urllib.parse

# Define a function for geocoding
def geocode_address(address):
    try:
        # Check if the address is missing or invalid
        if pd.isna(address) or address.strip() == '':
            return ("Address Missing", "Address Missing")

        # Encode the address to handle special characters
        encoded_address = urllib.parse.quote_plus(address)

        geocode_result = gmaps.geocode(encoded_address)

        if geocode_result:
            location = geocode_result[0]['geometry']['location']
            lat, lon = location['lat'], location['lng']
            return (lat, lon)
        else:
            return ("Not Found", "Not Found")
    except Exception as e:
        print(f"Error geocoding address '{address}': {str(e)}")
        return ("Error", "Error")

# Split your DataFrame into smaller chunks for concurrent processing
chunk_size = 100  # You can adjust this based on your available resources
chunks = [df[i:i + chunk_size] for i in range(0, len(df), chunk_size)]

# Initialize a list to store coordinates
coordinates = []

# Function to process a chunk of data
def process_chunk(chunk):
    for _, row in chunk.iterrows():
        coordinates.append(geocode_address(row['raw_address']))

# Start concurrent processing
with concurrent.futures.ThreadPoolExecutor() as executor:
    start_time = time.time()
    executor.map(process_chunk, chunks)
    end_time = time.time()

print(f"Geocoding completed in {end_time - start_time} seconds")

# Print the list of coordinates
print(coordinates)
```
